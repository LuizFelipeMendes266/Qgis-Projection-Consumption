# Packages

```python
import pandas as pd
import geopandas as gpd
import numpy as np
from unidecode import unidecode
```

# Core

To perform the projection, it is necessary to enter the IBGE code of the city. In this case, Campinas City.

```python

cod_muni = '3509502'
```

```python

df = pd.read_parquet('../data/prediction-expense-by-census-tract.parquet')
```

```python
df
```

Selecting the city

```python
df = df[df.code_muni == cod_muni]
df
```

```python
df.name_muni.unique()
```

We loaded the shape data for the state of the analyzed city, obtained from IBGE website.

```python

df_shape = gpd.read_file('35SEE250GC_SIR.shp')
```

```python
df_shape
```

We filter for the city of Campinas

```python

df_cidades = df_shape[df_shape['CD_GEOCODM'].isin([cod_muni])]
```

```python

df_cidades['NM_MUNICIP'].unique()
```

```python
df_cidades.head()
```

We combine the data from IBGE Census with our expenditure projection.

```python

df_geo = df.merge(df_cidades[['CD_GEOCODI','geometry']],how='left',left_on='id_setor_censitario',right_on='CD_GEOCODI')
df_geo
```

Since we have more than one income data, we perform the sum of the columns to generate the number of households.

```python

df_geo['numero_domicilio'] = df_geo.loc[:,['renda1','renda2','renda3']].sum(axis=1)
df_geo
```

Therefore, we just need to divide the total expenditure per sector by the number of households.

```python

df_geo['gasto_medio_setor'] = df_geo['gasto_total']/df_geo['numero_domicilio']
df_geo
```

Transfromation to a GeoPandas Dataframe

```python

df_geo = gpd.GeoDataFrame(df_geo)
```

```python
df_geo.name_muni.unique()
```

# Results

```python
df_geo.head()
```

Code to save the file in gpkg

```python


for i in df_geo.name_muni.unique():
    nome_cidade = str.lower(unidecode(i)).replace(' ','_')
    nome_arquivo = "setores_%s"
    nome_arquivo = nome_arquivo % nome_cidade + '.gpkg'
    df_geo[df_geo['name_muni']== i].to_file(nome_arquivo,driver='GPKG')
```
