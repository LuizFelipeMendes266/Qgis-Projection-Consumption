# Core

```python
df = pd.read_parquet('../data/campinas_iptu.parquet')
df
```

We select only the occurrences with terrain_area

```python

df = df[~df.terrain_area.isna()]
```

It is necessary to remove the occurrences that do not have coordinate data

```python
df = df[df['location_point'].map(lambda d: len(d)) > 0]
```

Since we have the coordinates, we can separate them into columns and create the 'geometry' column, which is the geometric union in GeoPandas of the coordinates.

```python


df['lat'] = [ i[1] for i  in df['location_point']]
df['lon'] = [ i[0] for i  in df['location_point']]
df['geometry'] = gpd.points_from_xy(df['lon'],df['lat'])
```

```python
df
```

Cleaning the dirty in the "owners" column.

```python
df['owners'] = df.owners.astype('str')
df['owners'] = [ re.findall(r"'name': '(.*?)'",i) for i  in df['owners']]
df['owners'] = [str(i)[2:-2] for i in df.owners]
df.head()
```

```python
df.use_type.unique()
```

```python
df.shape
```

Here, we only select the residential ocurrences

```python
# selecionar o tipo de uso desejado

df = df[df.use_type.isin(['Residencial Horizontal','Residencial Vertical'])]
df
```

Loading the shape file from IBGE

```python


df_shape = gpd.read_file('35SEE250GC_SIR.shp').to_crs(epsg='4326')
df_shape
```

Considering Campinas City, we have

```python

cod_muni = ['3509502']
```

```python
# filtragem do  cod_muni no df_shape
df_malha = df_shape[df_shape['CD_GEOCODM'].isin(cod_muni)]
```

df_malha = Dataset Shape from Campinas

```python
# Transforming to a GeoDataFrame

df = gpd.GeoDataFrame(df)
```

Performing the spacial merge between the two datasets

```python
# join entre o df do(s) município(os) e o df-shp filtrado

df_join = df.sjoin(df_malha[['CD_GEOCODI','geometry']],op='within',how='left')
df_join
```

```python
# exclusão do index criado pelo join
del df_join['index_right']
```

```python
df_join
```

Loading the prediction expense IPC.

```python

df_exp = pd.read_parquet('../data/prediction-expense-by-census-tract.parquet')
df_exp
```

Campinas Code = 3509502

```python
df_exp = df_exp[df_exp.code_muni == '3509502']
df_exp
```

Here, we have a brief aggregation to  obtain the mean,sum and the number of households from Campinas.

```python
# agrupamento com base no GEOCODI em vista de obter informações de área do df_join

info_por_setor = df_join.groupby(['CD_GEOCODI']).agg(
    area_terreno = ('terrain_area',sum),
    media_terreno = ('terrain_area','median'),
    unidades = ('realty_registration','count')

    
)
```

```python
info_por_setor
```

```python
# reset para normalizar index

info_por_setor.reset_index(inplace=True)
```

```python
info_por_setor
```

Since we have more than one income data, we perform the sum of the columns to generate the number of households.

```python
# soma das rendas e criação de nova coluna "numero_domicilio"
# seleção das colunas que interessam

df_exp['numero_domicilio'] = df_exp.loc[:,['renda1','renda2','renda3']].sum(axis=1)
df_exp = df_exp[['id_setor_censitario','renda1','renda2','renda3','numero_domicilio','gasto_total']]
```

```python
df_exp
```

So, now we can merge all the information

```python


df_merge = info_por_setor.merge(df_exp,how='left',right_on='id_setor_censitario',left_on='CD_GEOCODI')
df_merge
```

We divide the total expenditure by the number of households.

```python
# criação da coluna de gasto médio 

df_merge['gasto_medio_setor'] = df_merge['gasto_total'] / df_merge['numero_domicilio']
```

Removing NaN values

```python
# filtragem em vista de excluir nulos

df_merge = df_merge[~df_merge.gasto_medio_setor.isna()]
```

```python
df_merge
```

```python
# seleção de colunas

df_merge = df_merge[['CD_GEOCODI','media_terreno','gasto_medio_setor','unidades']]
```

```python
df_merge
```

```python
df_join
```

Finally, we have created the final dataset. The column "household expenditure" is generated by taking the average of the land cost and the sector expenditure.

```python
# por fim, merge para se obter o dataset final

df_final = df_join.merge(df_merge,how='left')
```

```python
# criação de coluna gasto por domicilio com base nos dados de área e gasto do setor


df_final['gasto_por_domicilio'] = (df_final['terrain_area'] / df_final['media_terreno']) * df_final['gasto_medio_setor']
```

```python
df_final
```

```python
# verificação de média 

df_final.gasto_por_domicilio.median()
```

Now, we can separate by categories.

```python
# criação de categorias com base no gasto por domicilio

categorias = ['F','E','D','C','B','A','A+']
cortes_bins = [0,2500,5000,10000,25000,50000,100000,5000000]
```

```python
df_final['faixa'] = pd.cut(df_final.gasto_por_domicilio,bins=cortes_bins,labels=categorias)
```

```python
df_final
```

```python
# drop de colunas que n interessam

df_final = df_final.drop(columns=['owners','location_point'])
```

```python
# filtragem para retirar aqueles que possuem código nulo

df_final = df_final[~df_final['CD_GEOCODI'].isna()]
```

```python
# transformar a coluna faixa em string para poder salvaar

df_final.faixa = df_final.faixa.astype('str')
```

```python
df_final = gpd.GeoDataFrame(df_final)
```
